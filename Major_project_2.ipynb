Q2. Reverse a String Write a function that takes a string as input and returns the reversed
string without using built-in reverse functions.  

```python
def reverse_string(s):
    reversed_str = ""

    for char in s:
        reversed_str = char + reversed_str

    return reversed_str

# Example
text = input("Enter string: ")
print("Reversed:", reverse_string(text))
```

 Q3. Function and Logic Create a user-defined function that:
● Accepts a number as input
● Returns whether the number is a Palindrome number or not

```python
def is_palindrome(num):
    original = num
    reversed_num = 0

    while num > 0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num = num // 10

    if original == reversed_num:
        return "Palindrome"
    else:
        return "Not Palindrome"

# Example
number = int(input("Enter number: "))
print(is_palindrome(number))
```

Q4. NumPy Array Operations
● Create a NumPy array of integers from 1 to 20

● Reshape it into a 4 × 5 matrix

● Find:

○ Mean of each column

○ Maximum value of each row

```python
import numpy as np
arr = np.arange(1, 21)

matrix = arr.reshape(4, 5)
print("Matrix:\n", matrix)

col_mean = np.mean(matrix, axis=0)
print("Column Mean:", col_mean)

row_max = np.max(matrix, axis=1)
print("Row Max:", row_max)
```

Q8. Data Analysis Answer the following using Pandas:

● Find the top 5 product categories by average rating

● Identify the 5 most expensive products

```python
import pandas as pd

# Load dataset (make sure file is uploaded)
df = pd.read_csv("flipkart_com-ecommerce_sample.csv")


# Convert columns to numeric
df['product_rating'] = pd.to_numeric(df['product_rating'], errors='coerce')
df['retail_price'] = pd.to_numeric(df['retail_price'], errors='coerce')

# Remove missing values
df = df.dropna(subset=['product_rating', 'retail_price'])

# Remove duplicates
df = df.drop_duplicates()


top_categories = (
    df.groupby('product_category_tree')['product_rating']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

print("Top 5 Categories by Average Rating:\n")
print(top_categories)

top_expensive = (
    df.sort_values(by='retail_price', ascending=False)
    .head(5)
)

print("\nTop 5 Most Expensive Products:\n")
print(top_expensive[['product_name', 'retail_price']])
```

#Section D – Exploratory Data Analysis

Q9. Visualization Create the following plots:

● Distribution of product prices

● Count plot of product categories  


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("flipkart_com-ecommerce_sample.csv")

# Convert columns
df['retail_price'] = pd.to_numeric(df['retail_price'], errors='coerce')
df['product_rating'] = pd.to_numeric(df['product_rating'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['retail_price', 'product_category_tree'])
# Price Distribution

plt.figure()
plt.hist(df['retail_price'], bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# Category Count

plt.figure()
sns.countplot(x='product_category_tree', data=df)
```
